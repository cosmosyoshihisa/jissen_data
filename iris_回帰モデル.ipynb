iris_dataを使って解析

```r
data(iris)
PL <- iris$Petal.Length
PW <- iris$Petal.Width
SL <- iris$Sepal.Length
SW <- iris$Sepal.Width
SP <- iris$Species
#Sepal:がく片、Petal:花びら
```

単回帰

```r
lsr1 <- lm(PW~PL)
```

```r
summary(lsr1)
plot(PL,PW)
abline(lsr1)
```

重回帰分析

```r
lsr2 <- lm(PW~PL+SP)
```

```r
summary(lsr2)
plot(PL,PW)
#abline(lsr2)
```

```r
a0 <-coef(lsr2)[1]
a0
```

```r
a1 <- coef(lsr2)[1]+coef(lsr2)[3]
a1
```

```r
a2<- coef(lsr2)[1] +coef(lsr2)[4]
b <- coef(lsr2)[2]
```

```r
plot( PL,PW, type ="n")
points( PL[SP=="setosa"],PW[SP=="setosa"],col="blue")
points( PL[SP=="versicolor"], PW[SP=="versicolor"],col="red")
points( PL[SP=="virginica"],PW[SP=="virginica"],col="green3")
abline(a=a0, b, col="blue")
abline(a=a1, b, col="red")
abline(a=a2, b, col="green3")
abline(lsr1)
```

```r
summary(iris)
```

```r
lsr3 <- lm(PW~PL*SP)
summary(lsr3)
```

```r
c0 <- coef(lsr3)[1]
c1 <- coef(lsr3)[1] +coef(lsr3)[3]
c2 <- coef(lsr3)[1]+ coef(lsr3)[4]

d0 <- coef(lsr3)[2]
d1 <- coef(lsr3)[2]+ coef(lsr3)[5]
d2 <- coef(lsr3)[2]+coef(lsr3)[6]

plot( PL,PW, type ="n")
points( PL[SP=="setosa"],PW[SP=="setosa"],col="blue")
points( PL[SP=="versicolor"], PW[SP=="versicolor"],col="red")
points( PL[SP=="virginica"],PW[SP=="virginica"],col="green3")
abline(a=c0, b= d0, col="blue")
abline(a=c1, b= d1, col="red")
abline(a=c2, b= d2, col="green3")
abline(lsr1)
```

```r
lsr12 <- lm(PW~PL+SW+SL+SP) 
summary(lsr12)
```

課題：自分たちで回帰モデルを作る！！

```r
lsr4 <- lm(PW~PL*SP+SW)
summary(lsr4)
```

```r
lsr5 <- lm(PW~PL*SP+SL)
summary(lsr5)
```

```r
lsr6 <- lm(PW~PL*SP+SW+SL)
summary(lsr6)#*SPに対して+SPもしくは+SL単体よりも+SP+SLの方が相関が良い
#SPは+よりも*が相関が良い
```

```r
lsr65 <- lm(PW~PL*SP*SW+SL)
summary(lsr65)
```

```r
lsr66 <- lm(PW~PL*SP*SL+SW)
summary(lsr66)
```

```r
lsr7 <- lm(PW~PL*SP*SW*SL)
summary(lsr7)#SP,SW,SLすべて*で一番良い結果
#SW,SLどちらかを+に変えてしまうと相関が悪くなる。特にSWを+に変えた時の相関の落ち方が大きい
```

```r
SW_2 <- SW^2#がく片の幅を2倍
```

```r
SW
```

```r
SW_2
```

```r
lsr8 <- lm(PW~PL*SP*SW_2*SL)
summary(lsr8)#SWよりSW_2での相関の方が良かった
```

```r
SL_2 <- SL^2
```

```r
lsr9 <- lm(PW~PL*SP*SW*SL_2)
summary(lsr9)#SLよりSL_2での相関の方が良かった(ほんのちょっと)
```

```r
lsr10 <- lm(PW~PL*SP*SW_2*SL_2)
summary(lsr10)
```

```r
SA = SL*SW #SA:Sepal Area
SA
```

```r
lsr11 <- lm(PW~PL*SP*SA)
summary(lsr11)#面積は考えない方が良さそう（特徴量減るから？？）
```

```r
lsr13 <- lm(PW~PL*SP*SA*SL)
summary(lsr13)
```

```r
lsr14 <- lm(PW~PL*SP*SA*SW)
summary(lsr14)
```

```r
lsr15 <- lm(PW~PL*SP*SA*SW*SL)
summary(lsr15)#面積という特徴量をさらに増やすことで少し相関が上がった
```

```r
SL_100<- SL^10
SW_100<- SW^10
```

```r
lsr16 <- lm(PW~PL*SP*SL_100*SW_100)
summary(lsr16)#がく片の幅と長さを誇張しても意味ない
```

```r
lsr17 <- lm(PW~PL*SP*SA*SW_2*SL_2)
summary(lsr17)#面積という特徴量をさらに増やすことで少し相関が上がった
```
